```python
import pandas as pd
import csv
import sqlite3 
conn= sqlite3.connect("covid.db")
data.to_sql('covid', conn, if_exists='append', index = False)

data

```

```python
def covid(requete):
    df =pd.read_sql_query(requete, conn)
    print(pd.DataFrame(df))
df_covid=covid('SELECT * FROM covid WHERE location ="India" ')
df_covid
```

##  1. Nombre de pays

```python
df_country = covid('SELECT COUNT(DISTINCT(location)) as "nombre_de_pays" FROM covid')
df_country
```

##  2. Most entry

```python
most_entry = covid('SELECT date, new_cases FROM covid GROUP BY date ORDER BY date DESC')
most_entry
```

## 3. highest single day deaths per million

```python
country_case = covid("SELECT location, max(new_deaths_per_million), date FROM covid")
country_case

```

## 4. the deaths per million exceeded 1,000 in one day

```python
df_birthday = pd.read_sql_query('SELECT new_deaths_per_million, new_cases_per_million, continent, date FROM covid WHERE date = "2020-08-09" GROUP BY continent', conn)
df_birthday
```

```python
labels = df_birthday['continent']
taux_de_reussite_pourcentage_G = df_birthday['new_deaths_per_million']


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots(figsize=(12, 7))
rects1 = ax.bar(x , taux_de_reussite_pourcentage_G, width, label='y')


ax.set_ylabel('Morts par millions d habitant')
ax.set_xlabel('Continent')
ax.set_title('Nombre de morts par million d habitant le 9 aout 2020  ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)


fig.tight_layout()

plt.show()
```
